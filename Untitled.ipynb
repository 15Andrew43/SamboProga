```python
N = len(DF)
for i in [1, 2, 4, 8, 16, 32, 64, 128, 256]:
    if i > N:
        add = i-N
        BinN = i
        break
add
```

```python
N = len(DF)
for i in range(add):
    DF.loc[N+i, 'Фио'] = '* *'
    DF.loc[N+i, 'Клуб'] = '***'
DF
```

```python
def proga(start, n):
    if n==2:
        return
    clubs = dict()
    for i in range(start, start+n):
        clubs.setdefault(DF['Клуб'][i], [0])
        clubs[DF['Клуб'][i]][0]+=1
        clubs[DF['Клуб'][i]].append(DF['Фио'][i])
    flag = 0 # 0 - need add people to left, 1 - need add people to right
    l = start
    r = start + n//2
    
    
    for i in clubs.keys():
        if clubs[i][0]%2 == 0:
            for j in range(1, 1 + clubs[i][0]//2):
                DF['Фио'][l] = clubs[i][j]
                DF['Клуб'][l] = i; 
                l+=1
            for j in range(1 + clubs[i][0]//2, clubs[i][0]+1):
                DF['Фио'][r] = clubs[i][j]
                DF['Клуб'][r] = i; 
                r+=1
        elif flag == 0:
            for j in range(1, 1 + clubs[i][0]//2 + 1):
                DF['Фио'][l] = clubs[i][j]
                DF['Клуб'][l] = i; 
                l+=1
            for j in range(1 + clubs[i][0]//2 + 1, clubs[i][0]+1):
                DF['Фио'][r] = clubs[i][j]
                DF['Клуб'][r] = i; 
                r+=1  
            flag = 1
        elif flag == 1:
            for j in range(1, 1 + clubs[i][0]//2):
                DF['Фио'][l] = clubs[i][j]
                DF['Клуб'][l] = i; 
                l+=1
            for j in range(1 + clubs[i][0]//2, clubs[i][0]+1):
                DF['Фио'][r] = clubs[i][j]
                DF['Клуб'][r] = i; 
                r+=1
            flag = 0

    proga(start, n//2)
    #print(DF)
    proga(start+n//2, n//2)
    
```
